### Creacion de sub Datasets, y Seleccion de columnas

```python
## Sub dataset copy()
## Sirve para copiar un dataset
movies_copy = movies.copy()
movies_copy

## Sub dataset loc[]
## Sirve para seleccionar datos por etiquetas
movies_copy.loc[2]

## Sub dataset loc[]
## Sirve para seleccionar datos por etiquetas
movies.loc[0:5]
```

### Filtrar Data

```python
## Filtrar por Genero
## BOOLEAN INDEXING
movies['Genre'] == 'Action'

## Filtrar por Genero con la data
movies[movies['Genre'] == 'Comedy']

## Filtrar por Genero con la data y por año
movies[(movies['Genre'] == 'Comedy')&(movies['Year'] == 2008)]

## Filtrar por Genero con la data y por año
movies[(movies['Genre'] == 'Comedy')|(movies['Genre'] == 'Romance')]

## Filtrar por el rate de Audience score arriba del 80%
movies[(movies['Audience score %'] >= 80)]

## Filtrar por el rate de Rotten Tomatoes arriba del 90%
movies[(movies['Rotten Tomatoes %'] >= 90)]

movies[(movies['Profitability'] >= 4)]
```

### Seleccion de Columnas

```python
## Seleccion de una columna
movies['Genre']

## Seleccion de Multiples Columnas
movies[['Genre','Lead Studio']]

## Identificar Columnas Nulas
movies.isnull()

## Identificar datos nulos en una columna
movies['Genre'].isnull()
movies[movies['Genre'].isnull()]

## Identificar datos NO nulos en una columna
movies['Genre'].notnull()
movies[movies['Genre'].notnull()]
```

### Ordenamiento de datos

```python
## Sort
movies.sort_values(by='Film')

## Sort
movies.sort_values(by='Film', ascending=False)

## Sort
movies.sort_values(by='Film', ascending=False)

## Sort
movies.sort_values(by='Film', ascending=True)

## Sort
movies.sort_values(by=['Film', 'Year'], ascending=[True, False])

## Sort
movies.sort_values(by=['Film', 'Year'], ascending=[False, True])


## Sort
# movies.sort_index()
```

## Iniciacion en Analitica de Datos

```python
## nlargest
## Sirve para seleccionar los n registros más grandes
movies.nlargest(5, 'Rotten Tomatoes %')

## nsmallest
## Sirve para seleccionar los n registros más pequeños
movies.nsmallest(5, 'Rotten Tomatoes %')
```
